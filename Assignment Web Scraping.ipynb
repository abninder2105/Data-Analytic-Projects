```python
# Que 3
din_out= requests.get("https://www.dineout.co.in/delhi-restaurants/welcome-back")
din_out
```

```python
soup1 = BeautifulSoup(din_out.content)
```

```python
soup1
```

```python
soup1.text
```

```python
restname = soup1.find('div',class_="restnt-info cursor")
restname
```

```python
titles = []

for i in soup1.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
    
print(titles)    
```

```python
locations = []
for i in soup1.find_all('div', class_="restnt-loc ellipsis"):
    locations.append(i.text)
    
locations    
```

```python
prices = []
for i in soup1.find_all('span', class_="double-line-ellipsis"):
    prices.append(i.text)

prices
```

```python
imgurl = []
for i in soup1.find_all('img', class_="no-img"):
    imgurl.append(i["data-src"])
    
imgurl
```

```python
ratings = []
for i in soup1.find_all("div", class_="restnt-rating rating-4"):
    ratings.append(i.text)
ratings.append('5')    

print(ratings)





```

```python



df = pd.DataFrame({'NAME':titles,'LOCATION':locations,'PRICE/CUISINES':prices,'IMG_URL':imgurl,'RATINGS': ratings})
df

```

```python
print(len(titles),len(locations),len(prices),len(imgurl),len(ratings))
```
